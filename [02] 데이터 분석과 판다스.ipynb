# ì¤‘ë³µ ë ˆì´ë¸”ì´ ìˆëŠ” Series

- Seriesì˜ ì¸ë±ìŠ¤ëŠ” ê¼­ ê³ ìœ (unique)í•  í•„ìš”ëŠ” ì—†ìŒ  
- ë™ì¼í•œ ì´ë¦„(ë ˆì´ë¸”)ì´ ì—¬ëŸ¬ ë²ˆ ë“±ì¥í•  ìˆ˜ë„ ìˆìŒ  
- ì´ ê²½ìš° `iloc`ê³¼ `loc`ì˜ ë™ì‘ì´ ë‹¬ë¼ì§

ğŸ‘‰ ì£¼ì˜!  
- **iloc** â†’ ìœ„ì¹˜ ê¸°ë°˜ â†’ í•˜ë‚˜ì˜ ê°’ ë°˜í™˜  
- **loc** â†’ ì´ë¦„ ê¸°ë°˜ â†’ ê°™ì€ ì´ë¦„ì´ ì—¬ëŸ¬ ê°œë©´ **ì—¬ëŸ¬ ê°’ ë°˜í™˜**

```python
import pandas as pd

age = pd.Series([25, 34, 19, 45, 60])
age.index = ['John', 'Jane', 'Tom', 'Micle', 'Tom']

print("Series ë‚´ìš©:")
print(age)

# ìœ„ì¹˜ ê¸°ë°˜ (iloc)
print("\nage.iloc[3] â†’")
print(age.iloc[3])

# ë ˆì´ë¸” ê¸°ë°˜ (loc)
print("\nage.loc['Tom'] â†’")
print(age.loc['Tom'])
```

# ìˆ«ì ë ˆì´ë¸” ì§€ì •í•˜ê¸°

- Seriesì˜ ì¸ë±ìŠ¤ëŠ” ê¼­ ë¬¸ìì—´ì´ ì•„ë‹ˆë¼ **ìˆ«ì**ë¡œë„ ì§€ì •í•  ìˆ˜ ìˆìŒ  
- í•˜ì§€ë§Œ ì£¼ì˜í•  ì :  
  - `iloc[ìˆ«ì]` â†’ **ìœ„ì¹˜ ê¸°ë°˜** (0ë¶€í„° ì‹œì‘í•˜ëŠ” ìˆœì„œ)  
  - `loc[ìˆ«ì]` â†’ **ë ˆì´ë¸” ê¸°ë°˜** (ì§ì ‘ ì§€ì •í•œ ì¸ë±ìŠ¤ ê°’)

ğŸ‘‰ ë”°ë¼ì„œ ìˆ«ì ë ˆì´ë¸”ì„ ì“°ë©´ í—·ê°ˆë¦´ ìˆ˜ ìˆìŒ
